```python
import random
import pyttsx3
import wikipedia
import time
import webbrowser
from datetime import datetime
import nltk



engine = pyttsx3.init()
now = datetime.now()
print("Hi! I am at your service. You can ask me anything you want.")
engine.say("i am at your service. You can ask me anything you want.")
engine.runAndWait()
greetings = ['hello','hey!','heya!','hi','wassup?','hola','hey']
questions = ['how are you?','how are you doing today?','how is everything going on?']
validations = ['yeah','yes','ofcourse','absolutely','NO','nah!']
responses = ['i am good','i am okay','im doing good too!','i am fine']
verifications = ['sure?','are you sure?','is it?']
var1 = ['who made you', 'who created you']
var2 = ['I was created by Edward right in his computer.', 'Edward', 'Some guy whom i never got to know.']
var3 = ['what time is it', 'what is the time', 'time','date and time']
var4 = ['who are you', 'what is you name','what is your name?','what are you called']
cmd1 = ['open browser', 'open google','google search']
cmd2 = ['play music', 'play songs', 'play a song','open youtube','i want to watch a video']
cmd3 = ['tell a joke', 'tell me a joke', 'say something funny', 'tell something funny']
jokes = ['Can a kangaroo jump higher than a house? Of course, a house doesnâ€™t jump at all.', 'My dog used to chase people on a bike a lot. It got so bad, finally I had to take his bike away.', 'Doctor: Im sorry but you suffer from a terminal illness and have only 10 to live.Patient: What do you mean, 10? 10 what? Months? Weeks?!"Doctor: Nine.']
cmd4 = ['technical events', 'upcoming tech events in India']
cmd5 = ['upcoming events at bvrit']
cmd6 = [ 'bye', 'goodbye','c ya!']
cmd7 = ['red', 'green', 'yellow', 'black','pink','violet','purple','orange']
colrep =['can you help me out with the details of a faculty memember','faculty details']
cmd8 = ['what is you favourite colour', 'what is your favourite color','your colour?']
#cmd9 = ['thank you']
repfr9 = ['youre welcome', 'glad i could help you']
var6 = ['online library','books','novels']
ser = ['wikipedia','information from wikipedia']
stu = ['student details']
var10 =['thank you!','thanks','thank you']
cmd10 = ['okay!','goodbye!','see you around!','until next time!','bye!']
while True:
    uinput = input()
    if uinput in greetings:
        print(random.choice(greetings)+" what is your good name?")
        engine.say(random.choice(greetings)+" what is your good name?")
        engine.runAndWait()
        name = input()
        print("Hello! "+name+ ", it's nice to meet you!")
        engine.say("Hello! "+name+ ", it's nice to meet you!")
        engine.runAndWait()
        print("How may I help you?")
        engine.say("How may I help you?")
        engine.runAndWait()
    elif uinput in questions:
        print(random.choice(responses))
        engine.say(random.choice(responses))
        engine.runAndWait()
    elif uinput in verifications:
        print(random.choice(validations))
        engine.say(random.choice(validations))
        engine.runAndWait()
    elif uinput in var1:
        print(random.choice(var2))
        engine.say(random.choice(var2))
        engine.runAndWait()
    elif uinput in var3:
        print(time.strftime("The time is %H:%M")+ now.strftime(" and The date is %m/%d/%y"))
        engine.say(time.strftime("The time is %H:%M")+ now.strftime(" and The date is %m/%d/%y"))
        engine.runAndWait()
    elif uinput in var4:
        print('i am Drake')
        engine.say('i am Drake')
        engine.runAndWait()
    elif uinput in cmd1:
        print('please wait...')
        engine.say('please wait...')
        engine.runAndWait()
        google = input('Google Search:')
        webbrowser.open_new_tab('http://www.google.com/search?btnG=1&q=%s' % google)
    elif uinput in cmd2:
        print('search your favourite songs on youtube!')
        engine.say('search your favourite videos on youtube!')
        engine.runAndWait()
        youtube = input('YouTube:')
        webbrowser.open_new_tab('https://www.youtube.com/%s' %youtube)        
    elif uinput in cmd3:
        print(random.choice(jokes))
        engine.say(random.choice(jokes))
        engine.runAndWait()
    elif uinput in colrep:
        print("sure, please wait for a second..")
        engine.say("sure, please wait for second..")
        engine.runAndWait()
        #department = ['cse','bsandh','ece','mech','civil','chem']
        engine.say('let me know the department:')
        engine.runAndWait()
        branch = input('let me know the department:')
        print('here are the required details')
        engine.say('here are the required details')
        engine.runAndWait()
        if branch == 'cse':
            webbrowser.open('http://www.bvrit.ac.in/index.php/cse-faculty')
        elif branch == 'bsandh':
            webbrowser.open('http://www.bvrit.ac.in/index.php/bsh-faculty')
        elif branch == 'chem':
            webbrowser.open('http://www.bvrit.ac.in/index.php/chemical-faculty')
        elif branch == 'ece':
            webbrowser.open('http://www.bvrit.ac.in/index.php/eee-faculty')
        elif branch == 'civil':
            webbrowser.open('http://bvrit.ac.in/index.php/civil-engineering-faculty')
        elif branch == 'IT':
            webbrowser.open('http://www.bvrit.ac.in/index.php/it-faculty')
        else:
            webbrowser.open('http://www.bvrit.ac.in/index.php/mech-faculty')
                
    elif uinput in cmd8:
        print('My favourite colour is: '+ random.choice(cmd7))
        engine.say('My favourite colour is: '+ random.choice(cmd7))
        print('Although it changes for every micro second!')
        engine.runAndWait()
        engine.say('Although it changes for every micro second!')
        engine.runAndWait()
    elif uinput in cmd4:
        print('here are a few tech events happening around the date..')
        engine.say('here are a few tech events happening around the date..')
        engine.runAndWait()
        webbrowser.open('https://10times.com/india/technology')
    elif uinput in cmd5:
        webbrowser.open('http://www.bvrit.ac.in/index.php/upcomming-events-menu?placement-details=')
    elif uinput in var10:
        print(random.choice(repfr9))
        engine.say(random.choice(repfr9))
        engine.runAndWait()
    elif uinput in cmd6:
        #cmd10 = ['okay!','goodbye!','see you around!','until next time!','bye!']
        print(random.choice(cmd10))
        engine.say(random.choice(cmd10))
        engine.runAndWait()
        exit()
    elif uinput in stu:
        engine.say("Here's the required student information")
        engine.runAndWait()
        webbrowser.open('https://bvrit.edu.in/StudentMaster.aspx')
    elif uinput in ser:
        a = input()
        print(wikipedia.summary(a,sentences=2))
        engine.say(wikipedia.summary(a,sentences=2))
        engine.runAndWait()
    elif uinput in var6:
        print("here's an online library for ya!")
        engine.say("here's an online library for ya!")
        engine.runAndWait()
        webbrowser.open('https://ndl.iitkgp.ac.in/result?q={%22t%22:%22search%22,%22k%22:%22cse%22,%22s%22:[],%22b%22:{%22filters%22:[]}}')
    else:
        print('hmm...guess you are busy..goodbye, see you later!')
        engine.say('hmm...guess you are busy..goodbye, see you later!')
        engine.runAndWait()
exit()
        
        
        
    

    

```
